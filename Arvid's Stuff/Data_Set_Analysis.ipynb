# Data Set Analysis

## Module Import

```python
# Data manipulation
import pandas as pd

# Math
import numpy as np 

# Visualization
import matplotlib.pyplot as plt
```

## Data Preparation

```python
# Read in data from a CSV file.
data = pd.read_csv('training_data_ht2025.csv')

# Make output numerical.
data["increase_stock_num"] = np.where(data['increase_stock']=='low_bike_demand', -1, 1)
data = data.drop('increase_stock', axis=1)
```

## Hour of Day Analysis

```python
# Hours of the day.
hours = range(24)

# Initialize an empty list to store the results for each hour.
hour_result = []

# Loop through the hours of the day.
for hour in hours:
    # Create a temporary data set with all entries at the specific hour.
    data_hour = data.loc[data['hour_of_day'] == hour]

    # Create a temporary data set with all entries at the specific hour and with increased demand.
    data_hour_increase_stock = data_hour.loc[data_hour['increase_stock_num'] == 1]

    # Save the share of entries where the stock was incresed.
    hour_result.append(len(data_hour_increase_stock)/len(data_hour))

# Bar plot to visualize the results.
plt.bar(hours, hour_result)
plt.xticks(np.arange(0,24, 2))
plt.xlabel("hour_of_day")
plt.ylabel("Share of entries with increased stock")
plt.show()
```

## Day of Week Analysis

```python
# Days of the week.
days = range(7)

# Initialize an empty list to store the results for each day.
day_result = []

# Loop through the days of the week.
for day in days:
    # Create a temporary data set with all entries at the specific day.
    data_day = data.loc[data['day_of_week'] == day]

    # Create a temporary data set with all entries at the specific day and with increased demand.
    data_day_increase_stock = data_day.loc[data_day['increase_stock_num'] == 1]

    # Save the share of entries where the stock was incresed.
    day_result.append(len(data_day_increase_stock)/len(data_day))

# Bar plot to visualize the results.
plt.bar(days, day_result)
plt.xticks(np.arange(0,7, 1))
plt.xlabel("day_of_week")
plt.ylabel("Share of entries with increased stock")
plt.show()
```

## Month of Year Analysis 

```python
# Months of the year.
months = range(1,13)

# Initialize an empty list to store the results for each month.
month_result = []

# Loop through the months of the year.
for month in months:
    # Create a temporary data set with all entries at the specific month.
    data_month = data.loc[data['month'] == month]

    # Create a temporary data set with all entries at the specific month and with increased demand.
    data_month_increase_stock = data_month.loc[data_month['increase_stock_num'] == 1]

    # Save the share of entries where the stock was incresed.
    month_result.append(len(data_month_increase_stock)/len(data_month))

# Bar plot to visualize the results.
plt.bar(months, month_result)
plt.xticks(np.arange(1,13, 1))
plt.xlabel("month")
plt.ylabel("Share of entries with increased stock")
plt.show()
```

## Weekday Analysis

```python
# Create two data sets with all entries at weekdays and weekends.
data_weekday = data.loc[data['weekday'] == 1]
data_weekend = data.loc[data['weekday'] == 0]

# Create two data sets with all entries at weekdays and weekends and with increased demand.
data_weekday_increase_stock = data_weekday.loc[data_weekday['increase_stock_num'] == 1]
data_weekend_increase_stock = data_weekend.loc[data_weekend['increase_stock_num'] == 1]

# Save the share of entries where the stock was incresed.
weekday_result = len(data_weekday_increase_stock)/len(data_weekday)
weekend_result = len(data_weekend_increase_stock)/len(data_weekend)

# Bar plot to visualize the results.
plt.bar(['weekday', 'weekend'], [weekday_result, weekend_result])
plt.ylabel("Share of entries with increased stock")
plt.show()
```

## Holiday Analysis

```python
# Create two data set with all entries at holidays and non-holidays.
data_holiday = data.loc[data['holiday'] == 1]
data_non_holiday = data.loc[data['holiday'] == 0]

# Create two data set with all entries at holidays and non-holidays and with increased demand.
data_holiday_increase_stock = data_holiday.loc[data_holiday['increase_stock_num'] == 1]
data_non_holiday_increase_stock = data_non_holiday.loc[data_non_holiday['increase_stock_num'] == 1]

# Save the share of entries where the stock was incresed.
holiday_result = len(data_holiday_increase_stock)/len(data_holiday)
non_holiday_result = len(data_non_holiday_increase_stock)/len(data_non_holiday)

# Bar plot to visualize the results.
plt.bar(['holiday', 'non-holiday'], [holiday_result, non_holiday_result])
plt.ylabel("Share of entries with increased stock")
plt.show()
```

## Temperature analysis

```python
# Create data sets with all entries at different temperature intervals.
data_temp1 = data.loc[(data['temp'] >= -10) & (data['temp'] < 0)]
data_temp2 = data.loc[(data['temp'] >= 0) & (data['temp'] < 10)]
data_temp3 = data.loc[(data['temp'] >= 10) & (data['temp'] < 20)]
data_temp4 = data.loc[(data['temp'] >= 20) & (data['temp'] < 30)]
data_temp5 = data.loc[(data['temp'] >= 30) & (data['temp'] < 40)]

# Create data sets with all entries at different temperature intervals and with increased demand.
data_temp1_increase_stock = data_temp1.loc[data_temp1['increase_stock_num'] == 1]
data_temp2_increase_stock = data_temp2.loc[data_temp2['increase_stock_num'] == 1]
data_temp3_increase_stock = data_temp3.loc[data_temp3['increase_stock_num'] == 1]
data_temp4_increase_stock = data_temp4.loc[data_temp4['increase_stock_num'] == 1]
data_temp5_increase_stock = data_temp5.loc[data_temp5['increase_stock_num'] == 1]

# Save the share of entries where the stock was incresed.
temp1_result = len(data_temp1_increase_stock)/len(data_temp1)
temp2_result = len(data_temp2_increase_stock)/len(data_temp2)
temp3_result = len(data_temp3_increase_stock)/len(data_temp3)
temp4_result = len(data_temp4_increase_stock)/len(data_temp4)
temp5_result = len(data_temp5_increase_stock)/len(data_temp5)

# Bar plot to visualize the results.
plt.bar(['-10 - -1', '0 - 9', '10 - 19', '20 - 29', '30 - 39'], [temp1_result, temp2_result, temp3_result, temp4_result, temp5_result])
plt.xlabel('temp [C]')
plt.ylabel("Share of entries with increased stock")
plt.show()
```

## Precipitation Analysis

```python
# Create data sets with all entries at different precipitation intervals.
data_precip1 = data.loc[(data['precip'] >= 0) & (data['precip'] < 10)]
data_precip2 = data.loc[(data['precip'] >= 10) & (data['precip'] < 20)]
data_precip3 = data.loc[(data['precip'] >= 20) & (data['precip'] < 30)]

# Create data sets with all entries at different precipitation intervals and with increased demand.
data_precip1_increase_stock = data_precip1.loc[data_precip1['increase_stock_num'] == 1]
data_precip2_increase_stock = data_precip2.loc[data_precip2['increase_stock_num'] == 1]
data_precip3_increase_stock = data_precip3.loc[data_precip3['increase_stock_num'] == 1]

# Save the share of entries where the stock was incresed.
precip1_result = len(data_precip1_increase_stock)/len(data_precip1)
precip2_result = len(data_precip2_increase_stock)/len(data_precip2)
precip3_result = len(data_precip3_increase_stock)/len(data_precip3)

# Bar plot to visualize the results.
plt.bar(['0 - 9', '10 - 19', '20 - 29'], [precip1_result, precip2_result, precip3_result])
plt.xlabel('precip [mm]')
plt.ylabel("Share of entries with increased stock")
plt.show()
```
